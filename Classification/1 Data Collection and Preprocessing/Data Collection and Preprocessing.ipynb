```python
import pandas as pd

Stage = gse.phenotype_data['characteristics_ch1.12.Stage']
Stage = Stage.rename("Stage")
Stage
```

```python
value_counts = Stage.value_counts()
value_counts
```

```python
data_matrix = gse.pivot_samples('VALUE')
data_matrix = data_matrix.append(Stage)
data_matrix_nf = data_matrix.dropna(subset=['Stage'], axis=1)
data_matrix_nf = data_matrix_nf.dropna()
data_matrix_nf = data_matrix_nf.T
print(data_matrix_nf)
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

data_matrix_nf["Stage"] = encoder.fit_transform(data_matrix_nf["Stage"])

data_matrix_nf.iloc[:,:-1]
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data_matrix_nf.iloc[:,:-1],data_matrix_nf.iloc[:,-1],test_size=0.35,random_state=42)
```

```python
import pickle

with open('lenc.pkl', 'wb') as f:
    pickle.dump((encoder), f)

with open('datacollected.pkl', 'wb') as f:
    pickle.dump((X_train, X_test, y_train, y_test), f)
```
