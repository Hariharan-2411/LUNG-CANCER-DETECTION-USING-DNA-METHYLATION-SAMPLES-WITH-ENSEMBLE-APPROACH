```python
import pandas as pd

recurrence = gse.phenotype_data['characteristics_ch1.13.recurrence']
recurrence = recurrence.rename("recurrence")
recurrence
```

```python
data_matrix = gse.pivot_samples('VALUE')
data_matrix = data_matrix.append(recurrence)
data_matrix_nf = data_matrix.dropna(subset=['recurrence'], axis=1)
data_matrix_nf = data_matrix_nf.dropna()
data_matrix_nf = data_matrix_nf.T
print(data_matrix_nf)
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

data_matrix_nf["recurrence"] = encoder.fit_transform(data_matrix_nf["recurrence"])

data_matrix_nf.iloc[:,:-1]
```

```python
data_matrix_nf.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data_matrix_nf.iloc[:,:-1],data_matrix_nf.iloc[:,-1],test_size=0.35,random_state=42)
```

```python
import pickle

with open('datacollected.pkl', 'wb') as f:
    pickle.dump((X_train, X_test, y_train, y_test), f)
```

```python
decoded_labels = encoder.inverse_transform(y_train)

decoded_labels = pd.DataFrame(decoded_labels)

vc = decoded_labels.value_counts()

vc
```
