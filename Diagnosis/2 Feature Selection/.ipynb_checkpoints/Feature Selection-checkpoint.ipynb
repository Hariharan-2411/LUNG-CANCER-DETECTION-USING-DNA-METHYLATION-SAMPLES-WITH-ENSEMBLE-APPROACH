```python
import pickle
with open('Dependency\datacollected.pkl', 'rb') as f:
    data_matrix_nf, test_data_matrix = pickle.load(f)
    
```

```python
from sklearn.linear_model import LassoCV
import pandas as pd

model = LassoCV(cv=5)
model.fit(data_matrix_nf.iloc[:,:-1], data_matrix_nf.iloc[:,-1])
features_select = []
# Print the list of features with non-zero coefficients
selected_features = list(data_matrix_nf.iloc[:,:-1].columns[model.coef_ != 0])
print("Selected features:", selected_features)
selected_features.extend(["tissue_char"])
features_select.append(selected_features)
```

```python
i = 0
j = 5260
counter = 0
while (j<=26300):
  model = LassoCV(cv=5)
  model.fit(data_matrix_nf.iloc[:,i:j], data_matrix_nf.iloc[:,-1])
  # Print the list of features with non-zero coefficients
  selected_features = list(data_matrix_nf.iloc[:,i:j].columns[model.coef_ != 0])
  selected_features.extend(["tissue_char"])
  features_select.append(selected_features)
  i = j + 1;
  j = j + 5260;
  counter = counter + 1
  print(counter," - ", len(selected_features))
print(features_select)
print(len(features_select))
```

```python
dm_reduced = []
for i in features_select:
    dm_reduced.append(data_matrix_nf[i])
dm_reduced
```

```python
testdm_reduced = []
for i in features_select:
    testdm_reduced.append(test_data_matrix[i])
testdm_reduced
```

```python
X = []
y = []
for i in dm_reduced:
    X.append(i.iloc[:,:-1])
    y.append(i.iloc[:,-1])


Xtest = []
Ytest = []
for i in testdm_reduced:
    Xtest.append(i.iloc[:,:-1])
    Ytest.append(i.iloc[:,-1])
```

```python
with open('FeatureSelectedData.pkl', 'wb') as f:
    pickle.dump((X, y, Xtest, Ytest), f)
```
