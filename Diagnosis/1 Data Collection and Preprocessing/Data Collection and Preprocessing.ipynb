```python
import GEOparse

gse = GEOparse.get_GEO(geo="GSE32861", destdir="/bin/LungCancer")
```

```python
print()
print("GSM example:")
for gsm_name, gsm in gse.gsms.items():
    print("Name: ", gsm_name)
    print("Metadata:",)
    for key, value in gsm.metadata.items():
        print(" - %s : %s" % (key, ", ".join(value)))
    print ("Table data:",)
    print (gsm.table.head())
    break

print()
print("GPL example:")
for gpl_name, gpl in gse.gpls.items():
    print("Name: ", gpl_name)
    print("Metadata:",)
    for key, value in gpl.metadata.items():
        print(" - %s : %s" % (key, ", ".join(value)))
    print("Table data:",)
    print(gpl.table.head())
    break
```

```python
import pandas as pd

tissue = gse.phenotype_data['characteristics_ch1.7.tissue']
tissue = tissue.rename("tissue_char")
```

```python
data_matrix = gse.pivot_samples('VALUE')
data_matrix = data_matrix.append(tissue)
data_matrix_nf = data_matrix.dropna()
data_matrix_nf = data_matrix_nf.T
print(data_matrix_nf)
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

data_matrix_nf["tissue_char"] = encoder.fit_transform(data_matrix_nf["tissue_char"])

data_matrix_nf.iloc[:,:-1]
```

```python
import pandas as pd

tissue_test = gse_test.phenotype_data['source_name_ch1']
tissue_test = tissue_test.rename("tissue_char")
```

```python
test_data_matrix = gse_test.pivot_samples('VALUE')
test_data_matrix = test_data_matrix.append(tissue_test)
test_data_matrix = test_data_matrix.transpose()
test_data_matrix = test_data_matrix.dropna()
test_data_matrix
```
